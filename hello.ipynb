## Etape 1

```java
import java.util.Scanner;

// Préparation de l'outil pour lire le clavier
Scanner clavier = new Scanner(System.in);
int nombre;

// La boucle commence ici
do {
    System.out.println("Veuillez saisir un nombre entre 2 et 15 : ");
    
    // On attend que l'utilisateur tape un nombre
    nombre = clavier.nextInt();
    
    // Si le nombre n'est pas valide, on affiche le message d'erreur
    if (nombre < 2 || nombre > 15) {
        System.out.println("Erreur : " + nombre + " n'est pas valide.");
    }

} while (nombre < 2 || nombre > 15); // On recommence tant que c'est invalide

System.out.println("Saisie validée : " + nombre);
```

## Etape 2

```java
// On demande le style à l'utilisateur
System.out.println("Voulez-vous un carré 'creux' ou 'plein' ? (tapez creux ou plein)");
String choix = clavier.next(); 

System.out.println("Voici votre carré de taille " + nombre + " :");

// BOUCLE 1 : Pour chaque ligne (de 0 à nombre-1)
for (int i = 0; i < nombre; i++) {
    
    // BOUCLE 2 : Pour chaque colonne dans la ligne (de 0 à nombre-1)
    for (int j = 0; j < nombre; j++) {
        
        // CAS 1 : L'utilisateur veut un carré PLEIN
        if (choix.equals("plein")) {
            System.out.print("* ");
        } 
        // CAS 2 : L'utilisateur veut un carré CREUX
        else {
            // On affiche une étoile SEULEMENT si on est sur un bord :
            // i==0 (première ligne) OU i==nombre-1 (dernière ligne)
            // j==0 (première colonne) OU j==nombre-1 (dernière colonne)
            if (i == 0 || i == nombre - 1 || j == 0 || j == nombre - 1) {
                System.out.print("* ");
            } else {
                // Sinon, on met deux espaces pour combler le vide
                System.out.print("  ");
            }
        }
    }
    // À la fin de chaque ligne (quand la boucle 2 est finie), on retourne à la ligne
    System.out.println();
}
```

## Etape 3

```java
System.out.println("Voici votre triangle de hauteur " + nombre + " :");

// Boucle principale : gère les lignes (de 0 à nombre-1)
for (int i = 0; i < nombre; i++) {
    
    // 1. D'abord, on affiche les ESPACES pour décaler le début
    // Plus on descend (i augmente), moins on met d'espaces
    for (int j = 0; j < nombre - 1 - i; j++) {
        System.out.print(" ");
    }
    
    // 2. Ensuite, on affiche les ÉTOILES
    // On affiche " * " (étoile + espace) pour faire une jolie pyramide
    for (int k = 0; k <= i; k++) {
        System.out.print("* ");
    }
    
    // 3. On passe à la ligne suivante
    System.out.println();
}
```

## Etape 4

```java
if (nombre <= 3) {
    // Cas simple (triangle normal code précédent)
    for (int i = 0; i < nombre; i++) {
        for (int j = 0; j < nombre - 1 - i; j++) System.out.print(" ");
        for (int k = 0; k <= i; k++) System.out.print("* ");
        System.out.println();
    }
} else {
    // Cas BONUS : Le Sapin (Triangles superposés)
    // On découpe en 3 étages pour imiter l'exemple
    System.out.println("Mode Sapin activé !");
    
    // On fait 3 blocs de triangles qui grandissent
    for (int bloc = 0; bloc < 3; bloc++) {
        
        // Chaque bloc a une hauteur de 3 lignes
        for (int i = 0; i < 3; i++) {
            
            // Calcul savant pour les espaces (pour centrer le sapin)
            // On se base sur la largeur max (nombre)
            for (int espace = 0; espace < nombre - 1 - (i + bloc); espace++) {
                System.out.print(" ");
            }
            
            // Calcul pour les étoiles
            for (int etoile = 0; etoile <= (i + bloc); etoile++) {
                System.out.print("* ");
            }
            System.out.println();
        }
    }
}
```

## Etape BONUS

```java
System.out.println("BONUS : Voici un diamant de taille " + nombre);

// PARTIE HAUTE (C'est exactement ton triangle)
for (int i = 0; i < nombre; i++) {
    // Espaces
    for (int j = 0; j < nombre - 1 - i; j++) {
        System.out.print(" ");
    }
    // Étoiles
    for (int k = 0; k <= i; k++) {
        System.out.print("* ");
    }
    System.out.println();
}

// PARTIE BASSE (Le triangle inversé)
// Attention : on commence à "nombre - 2" pour ne pas répéter la ligne du milieu deux fois
for (int i = nombre - 2; i >= 0; i--) {
    
    // Espaces (même logique)
    for (int j = 0; j < nombre - 1 - i; j++) {
        System.out.print(" ");
    }
    
    // Étoiles (même logique)
    for (int k = 0; k <= i; k++) {
        System.out.print("* ");
    }
    System.out.println();
}
```
